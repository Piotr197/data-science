```python
r = requests.get('https://www.spidersweb.pl/')
soup2 = BeautifulSoup(r.content, 'html.parser')
links = set()


for elem in soup.find_all(href=True):
    link = 'https://www.spidersweb.pl/' + elem.get('href')
    links.add(link)

row = 1
for link in links:
    linki.cell(row,column=1).value = link
    row +=1
```

```python
req = requests.get("https://www.filmweb.pl/Pulp.Fiction")
soup = BeautifulSoup(req.content, 'html.parser')

rezyser = soup.find(text="reżyseria").next_element.find("a").text
boxoffice = soup.find(text="boxoffice").next_element.text[1:].replace(" ", "")
ocena = float(soup.find("span", itemprop="ratingValue").text.replace(",", "."))
premiera = soup.find(text='premiera').next_element.find("span").text


filmweb.cell(1, 1).value = ocena
filmweb.cell(1, 2).value = rezyser
filmweb.cell(1, 3).value = boxoffice
filmweb.cell(1, 4).value = premiera


print(f"ocena: {ocena}, rezyseria:{rezyser}, boxoffice: {boxoffice}, premiera: {premiera}")
```
